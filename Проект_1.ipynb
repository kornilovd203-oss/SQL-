```python
options=csv_to_sqlite.CsvOptions(typing_style='full',encoding='utf-8')
input_name = ['funding_round.csv', 'acquisition.csv',
              'company.csv', 'people.csv', 'education.csv',
              'investment.csv', 'fund.csv']
database_name = 'output.sqlite'
if os.path.exists(database_name):
    os.remove(database_name)
csv_to_sqlite.write_csv(input_name, database_name, options)
```

```python
con=sqlite3.connect('output.sqlite')
```

```python
count_closed = pd.read_sql_query('''SELECT *
                                    FROM company
                                    WHERE status = 'closed';''',con)
count_closed
```

```python
sum_money_news = pd.read_sql_query('''SELECT SUM(funding_total_usd)
                                                 FROM company
                                                 WHERE category_code='news'
                                                 AND country_code='USA'
                                                 GROUP BY name
                                                 ORDER BY SUM(funding_total_usd) DESC;''',con)

sum_money_news
```

```python
silver_people = pd.read_sql_query ('''SELECT first_name,
                                    last_name,
                                    affiliation_name
                                    FROM people
                                    WHERE affiliation_name LIKE 'Silver%';''',
                                    con)
silver_people
```

```python
money_people = pd.read_sql_query ('''SELECT *
                                    FROM people
                                    WHERE last_name LIKE 'K%' AND affiliation_name LIKE '%money%';''',
                                    con)
money_people
```

```python
sum_country_money = pd.read_sql_query('''SELECT SUM(funding_total_usd), country_code
                                         FROM company
                                         GROUP BY country_code
                                         ORDER BY SUM(funding_total_usd) DESC;''',con)

sum_country_money
```

```python
min_max_invest = pd.read_sql_query ('''SELECT funded_at,
                                       MIN(raised_amount) AS minimum,
                                       MAX(raised_amount) AS maximum
                                       FROM funding_round
                                       GROUP BY funded_at
                                       HAVING (MIN(raised_amount)!=0
                                       AND MIN(raised_amount)!=MAX(raised_amount));''',
                                    con)
min_max_invest
```

```python
import pandas as pd

company['founded_at'] = pd.to_datetime(company['founded_at'])

company['year_founded_at'] = company['founded_at'].dt.year


number23 = company.query('(year_founded_at == 2011.0) | (year_founded_at == 2012.0) | (year_founded_at == 2013.0)').pivot_table(
    values = 'funding_total_usd',
    columns = 'year_founded_at',
    index = 'country_code',
    aggfunc = 'mean'
)

display(number23)
```
